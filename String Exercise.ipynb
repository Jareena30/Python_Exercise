```python
# Write a Python program to count the number of characters in a string.

str_demo = input("Enter String : ")  # input string
print("Original String : ", str_demo) # Original String
l = len (str_demo) #length of string
print("Length of String : ", l) #original string length
result = len([i for i in str_demo if i.isalpha()]) #total characters in string
print("Total number of characters in a String : ", result)
```

```python
# Write a Python program to check if a string is a palindrome.
str_demo = input("Enter String : ")  # input string
print("Original String : ", str_demo) # Original String
str_rev=str_demo[::-1] #reverse string
print("Reverse String :", str_rev)

if str_demo ==str_rev :   # condition to check whether string is palindrome
  print(str_demo,"and", str_rev, "is Palindrome") # true
else :
  print(str_demo,"and", str_rev, "is not Palindrome") # false
```

```python
# Write a Python program to remove all the vowels from a string.
str_demo = input("Enter String : ")  # input string
print("Original String : ", str_demo) # Original String
vowel = ['a','e','i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'] #vowels
result = [i for i in str_demo if i not in vowel] #checking for vowels
res = " " .join(result) # remaining letters except vowels
print(res)
```

```python
# Write a Python program to find the most frequent character in a string.
str_demo = input("Enter String : ")  # input string
print("Original String : ", str_demo) # Original String

freq = {}
for i in str_demo:
  if i in freq:
    freq[i] += 1
  else:
    freq[i] = 1
res = max(freq, key = freq.get)
print ("The most frequent character is", str(res))
```

```python
# Write a Python program to check if a string is a valid email address.
import re  # provides support for regular expressions

regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b'

def check(email):

  if(re.fullmatch(regex, email)):
	  print("You have entered valid Email")

  else:
    print("You have entered Invalid Email")

if __name__ == '__main__':
  email = input("Enter your Email : ")  # input string
  print("Entered Email is : ", email) # Original String
check(email)
```

```python
# Write a Python program to find the length of the longest substring without repeating characters.
def longest_substring_length(string):
    char_index = {}
    max_length = start = 0

    for i, char in enumerate(string):
        if char in char_index and char_index[char] >= start:
            start = char_index[char] + 1
        char_index[char] = i
        max_length = max(max_length, i - start + 1)

    return max_length

string = input("Enter a string: ")
length = longest_substring_length(string)
print("Length of longest substring without repeating characters:", length)
```
