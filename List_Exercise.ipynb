```python
#Write a Python program to find the sum of all elements in a list
listdata = []    #empty list
n = int(input("Enter number of elements : "))  # total no of elements in list

for i in range(0, n):    # elements
    e = int(input())
    listdata.append(e)   # appending elemebnts to the list

listdata     # list after adding elements

sum = 0
for ele in range(0, len(listdata)):
	sum = sum + listdata[ele]           # calculating sum of elements

print("Sum of all elements in given list: ", sum)
```

```python
# Write a Python program to check if a list is sorted in ascending order
listdata = []    #empty list
n = int(input("Enter number of elements : "))  # total no of elements in list

for i in range(0, n):    # elements
    e = int(input())
    listdata.append(e)   # appending elemebnts to the list

listdata     # list after adding elements
list1 = listdata[:]
list1.sort()  #sorted the list
print("List is Sorted in Ascending order : ", list1)
```

```python
# Write a Python program to find the common elements between two lists.
listdata = []    #empty list
n = int(input("Enter number of elements in list 1: "))  # total no of elements in list
for i in range(0, n):    # elements
    e = input()
    listdata.append(e)
listdata     # list1 after adding elements

listdata1 = []    #empty list
n = int(input("Enter number of elements in list 2: "))  # total no of elements in list
for i in range(0, n):    # elements
    e1 = input()
    listdata1.append(e1)
listdata1     # list2 after adding elements

def common_member(listdata, listdata1):
	a_set = set(listdata)
	b_set = set(listdata1)

	if (a_set & b_set):
		print("Common Elements :", a_set & b_set)
	else:
		print("No common elements")
common_member(listdata, listdata1)
```

```python
# Write a Python program to remove the nth occurrence of an element from a list.
listdata = []    #empty list
n = int(input("Enter number of elements : "))  # total no of elements in list
for i in range(0, n):    # elements
    e = input()
    listdata.append(e)
listdata     # list after adding elements

c=[]
count=0
b=input("Enter element to remove: ")
p=int(input("Enter the occurrence to remove: "))
for i in listdata:
    if(i==b):
        count=count+1
        if(count!=p):
            c.append(i)
    else:
        c.append(i)
if(count==0):
    print("Item not found ")
else:
    print("The number of repetitions is: ",count)
    print("Updated list is: ",c)
    print("The distinct elements are: ",set(listdata))
```

```python
# Write a Python program to remove the elements of a list that are divisible by 3.
listdata = []    #empty list
n = int(input("Enter number of elements : "))  # total no of elements in list
for i in range(0, n):    # elements
    e = input()
    listdata.append(e)
listdata        # list after adding elements

for i in listdata:
  if int(i) % 3 == 0:
    listdata.remove(i)
print("New list after removing elements: ", listdata)
```
