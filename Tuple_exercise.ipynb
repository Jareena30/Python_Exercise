```python
# Write a Python program to find the length of a tuple.
tuple_data = []

n = int(input ("Enter no of elements :"))
for i in range(n):    # elements
  tuple_data.append(int(input("Enter Element:")) )  # appending elemebnts to the list

tuple_data =tuple(tuple_data)   # converting list to tuple


print("Length of Tuple :", len(tuple_data))  #length of tuple
```

```python
# Write a Python program to concatenate two tuples.
tuple_data1 = []

n = int(input ("Enter no of elements for first list:"))
for i in range(n):    # elements
  tuple_data1.append(int(input("Enter Element:")) )  # appending elemebnts to the list

tuple_data1 =tuple(tuple_data1)   # converting list to tuple

tuple_data2 = []

n = int(input ("Enter no of elements for second list:"))
for i in range(n):    # elements
  tuple_data2.append(int(input("Enter Element:")) )  # appending elemebnts to the list

tuple_data2 =tuple(tuple_data2)   # converting list to tuple

tuple_data3 = tuple_data1 +tuple_data2    #concatenation

print(tuple_data3)
```

```python
# Write a Python program to count the number of occurrences of an element in a tupl
tup = []

n = int(input ("Enter no of elements :"))
for i in range(n):    # elements
  tup.append(int(input("Enter Element:")) )  # appending elemebnts to the list

tup =tuple(tup)   # converting list to tuple

def countX(tup, x):
	count = 0
	for ele in tup:
		if (ele == x):
			count = count + 1
	return count

enq = int(input ("Enter  elements to find no of occurance :"))

print(countX(tup, enq))
```

```python
# Write a Python program to find the maximum and minimum elements in a tuple.
tuple_data = (1,2,3,4,6,7,8)  #tuple
print('Tuple Data :', tuple_data)

list1  = list(tuple_data)
temp =  sorted(list1)
print('Maximum element is', temp[-1])

print('Minimum element is', temp[0])
```

```python
# Write a Python program to reverse a tuple.
tuple_data = (1,2,3,4,6,7,8)  #tuple
print('Tuple Data :', tuple_data)

list1  = list(tuple_data)
print('Reverse Tuple:', tuple(list1[::-1]))
```

```python
# Write a Python program to check if all elements in a tuple are the same.
my_tuple = (1, 1, 1, 1,)
reversed_tuple = my_tuple[::-1]
print("Reversed tuple:", reversed_tuple)
if my_tuple == reversed_tuple :
  print("All the elements in a tuple are same")
else :
  print("All the elements in a tuple are not same")
```
